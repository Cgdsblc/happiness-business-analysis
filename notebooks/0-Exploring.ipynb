```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "data/2019.csv"
happiness_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(happiness_data.head())

# Exclude non-numeric columns from correlation matrix
numeric_data = happiness_data.select_dtypes(include='number')

# Display the first few rows of the dataset
print(numeric_data.head())

# Descriptive statistics
print(numeric_data.describe())

# Correlation matrix
correlation_matrix = numeric_data.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
plt.title('Correlation Heatmap')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.show()


# Distribution of Happiness Scores
plt.figure(figsize=(10, 6))
sns.histplot(happiness_data['Score'], bins=20, kde=True)
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()

# Relationship between GDP per capita and Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP per capita', y='Score', data=happiness_data)
plt.title("Relationship between GDP per capita and Happiness Score")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.show()
```
