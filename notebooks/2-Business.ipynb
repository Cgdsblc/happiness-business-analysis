```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "data/2019.csv"
happiness_data = pd.read_csv(file_path)

# Select relevant columns for correlation analysis
selected_columns = ['GDP per capita', 'Social support', 'Healthy life expectancy',
                    'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Score']

# Create a subset dataframe with selected columns
subset_df = happiness_data[selected_columns]

# Calculate correlation matrix
correlation_matrix = subset_df.corr()

# Plot a heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Create a bar plot to visualize the distribution of happiness scores across regions
plt.figure(figsize=(14, 8))
sns.barplot(x='Score', y='Country or region', data=happiness_data, palette='viridis')
plt.title('Distribution of Happiness Scores Across Regions')
plt.xlabel('Happiness Score')
plt.ylabel('Country or Region')
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='GDP per capita', y='Score', data=happiness_data, color='blue', alpha=0.7)
plt.title('Relationship between GDP per Capita and Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = happiness_data['GDP per capita'].corr(happiness_data['Score'])
print(f'Correlation Coefficient: {correlation_coefficient}')
```

```python
# Load the datasets
df_2018 = pd.read_csv('data/2018.csv')
df_2019 = pd.read_csv('data/2019.csv')

# Select relevant columns
df_2018 = df_2018[['Country or region', 'Score']]
df_2019 = df_2019[['Country or region', 'Score']]

# Merge the datasets
combined_data = pd.merge(df_2018, df_2019, on='Country or region', suffixes=('_2018', '_2019'))

# Calculate the change in happiness score for each country
combined_data['Score Change'] = combined_data['Score_2019'] - combined_data['Score_2018']

# Line plot for average happiness score over the years
plt.figure(figsize=(12, 8))
sns.lineplot(x='Country or region', y='Score_2018', data=combined_data, label='2018', marker='o')
sns.lineplot(x='Country or region', y='Score_2019', data=combined_data, label='2019', marker='o')
plt.title('Average Happiness Score (2018-2019)')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=90)
plt.legend()
plt.show()

# Line plot for average happiness score change over the years
plt.figure(figsize=(12, 8))
sns.lineplot(x='Country or region', y='Score Change', data=combined_data, marker='o')
plt.title('Change in Happiness Score (2018-2019)')
plt.xlabel('Country')
plt.ylabel('Happiness Score Change')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df_2019 = pd.read_csv('data/2019.csv')

# Define a threshold for high and low happiness scores
happiness_threshold = df_2019['Score'].mean()  # You can adjust this threshold based on your criteria

# Create a new column to categorize countries into high and low happiness
df_2019['Happiness Category'] = df_2019['Score'].apply(lambda x: 'High' if x >= happiness_threshold else 'Low')

# Comparative analysis of GDP per capita for high and low happiness categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='Happiness Category', y='GDP per capita', data=df_2019)
plt.title('Comparative Analysis of GDP per Capita')
plt.xlabel('Happiness Category')
plt.ylabel('GDP per Capita')
plt.show()

# Comparative analysis of Social Support for high and low happiness categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='Happiness Category', y='Social support', data=df_2019)
plt.title('Comparative Analysis of Social Support')
plt.xlabel('Happiness Category')
plt.ylabel('Social Support')
plt.show()
```

```python
# Create a new column to categorize countries into high and low happiness
df_2019['Happiness Category'] = df_2019['Score'].apply(lambda x: 'High' if x >= happiness_threshold else 'Low')

# Factors for comparative analysis
factors = ['Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

# Loop through each factor for comparative analysis
for factor in factors:
    plt.figure(figsize=(12, 8))
    sns.boxplot(x='Happiness Category', y=factor, data=df_2019)
    plt.title(f'Comparative Analysis of {factor}')
    plt.xlabel('Happiness Category')
    plt.ylabel(factor)
    plt.show()
```
